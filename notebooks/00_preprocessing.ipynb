```python
import pandas as pd

try:
    df = pd.read_csv('../data/ndt_tests_corrigido.csv')
except FileNotFoundError:
    print("Erro: O arquivo '.csv' não foi encontrado. Verifique o caminho.")
```

```python
# Conversão da Coluna 'timestamp' para o tipo datetime
df['timestamp'] = pd.to_datetime(df['timestamp'],errors='coerce')
df['timestamp'] = df['timestamp'].dt.tz_localize(None)
```

```python
print("\n--- VALORES NULOS ---")
nulos = df.isnull().sum()
percentual = (nulos / len(df) * 100).round(2)

missing_info = pd.Series({
    col: f"{nulos[col]} ({percentual[col]}%)"
    for col in df.columns if nulos[col] > 0
}).sort_values(ascending=False)

if missing_info.empty:
    print("Nenhuma coluna contém valores nulos.")
else:
    print("Coluna | Total (Percentual)")
    print("-------|---------------------")
    for col, info in missing_info.items():
        print(f"{col:<6} | {info}")

print("\n--- LINHAS DUPLICADAS ---")
total_duplicadas = df.duplicated().sum()
print(f"Total: {total_duplicadas}")

print("\n--- VALORES ÚNICOS POR COLUNA (Cardinalidade) ---")
print(df.nunique().sort_values(ascending=False))
```
